```python
#defualt code
import pandas as pd


file_path = "your_file_path.xlsx"
df = pd.read_excel(file_path)
```

```python
#fucntion maker

def make_fltered_file(data_frame, keywords):
    
    df = data_frame.copy()
    keyword_masks = [df['공사명'].str.contains(keyword) for keyword in keywords]

    filtered_df = df[pd.concat(keyword_masks, axis=1).all(axis=1)]

    
    # Create a boolean mask for the rows to be dropped
    rows_to_drop = pd.concat(keyword_masks, axis=1).all(axis=1)

    # Drop the rows that meet the filter conditions
    df.drop(df.index[rows_to_drop], inplace=True)

    return([df,filtered_df])

def make_filtered_without_repetition(data_frame, keyword_list):
    df = data_frame  # Copy the original DataFrame to df
    print("original len:", len(df))  # Print the length of the original DataFrame

    for i in keyword_list:  # Iterate over the keywords in keyword_list
        
        tmp = make_fltered_file(df, i)  # Call a function to filter data based on the current keyword
        df = tmp[0]  # Update df with the modified DataFrame from the function
        filtered_df = tmp[1]  # Store the filtered DataFrame in filtered_df

        # Print the length of the current df (after filtering) and the keyword being used, 
        # as well as the length of the filtered_df
        print("len of df", len(df), "filter: ", i, "len of filtered_df", len(filtered_df))
        print('filtered_df.head()', filtered_df["공사명"].head())
        print("________")
        print("________")
    # Print the final length of df and the entire df DataFrame
    print(len(df))
    

    return(df["공사명"])




```
